```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras
from keras import layers
from tensorflow.keras import datasets, utils, layers, models
```

```python
(X_train, y_train), (X_test, y_test) = datasets.cifar10.load_data()
```

```python
NUM_CLASSES = 10

X_train = X_train.astype('float32')/255.0
X_test = X_test.astype('float32')/255.0

y_train = utils.to_categorical(y_train, NUM_CLASSES)
y_test = utils.to_categorical(y_test, NUM_CLASSES)
```

```python
input_layer = layers.Input(shape=(32,32,3))
x = layers.Flatten()(input_layer)
x = layers.Dense(units = 200, activation='relu')(x)
x = layers.Dense(units = 150, activation = 'relu')(x)
output_layer = layers.Dense(units=10, activation='softmax')(x)
model = models.Model(inputs = input_layer, outputs = output_layer)
```

```python
from keras.src.layers import Conv2D
conv_layer = layers.Conv2D(
    filters=32,
    kernel_size=3,
    strides = 1,
    padding = 'same',
    )(input_layer)
conv_layer = layers.BatchNormalization()(conv_layer)
conv_layer = layers.LeakyReLU()(conv_layer)

conv_layer = layers.Conv2D(
    filters=32,
    kernel_size=3,
    strides = 2,
    padding = 'same',
    )(input_layer)
conv_layer = layers.BatchNormalization()(conv_layer)
conv_layer = layers.LeakyReLU()(conv_layer)

conv_layer = layers.Conv2D(
    filters=64,
    kernel_size=3,
    strides = 1,
    padding = 'same',
    )(input_layer)
conv_layer = layers.BatchNormalization()(conv_layer)
conv_layer = layers.LeakyReLU()(conv_layer)

conv_layer = layers.Conv2D(
    filters=64,
    kernel_size=3,
    strides = 2,
    padding = 'same',
    )(input_layer)
conv_layer = layers.BatchNormalization()(conv_layer)
conv_layer = layers.LeakyReLU()(conv_layer)

conv_layer = layers.Flatten()(conv_layer)

conv_layer = layers.Dense(units = 128)(conv_layer)
conv_layer = layers.BatchNormalization()(conv_layer)
conv_layer = layers.LeakyReLU()(conv_layer)
conv_layer = layers.Dropout(0.5)(conv_layer)

conv_output_layer = layers.Dense(units = 10, activation = 'softmax')(conv_layer)

conv_model = models.Model(input_layer,conv_output_layer)
```

```python
model.summary()

conv_model.summary()
```

```python
from tensorflow.keras import optimizers
```

```python
opt = optimizers.Adam(learning_rate=0.0005)
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])

conv_opt = optimizers.Adam(learning_rate=0.0005)
conv_model.compile(loss='categorical_crossentropy', optimizer=conv_opt, metrics=['accuracy'])
```

```python
model.fit(
    X_train,
    y_train,
    batch_size=32,
    epochs=10,
    shuffle=True
)
```

```python
model.evaluate(X_test, y_test)
```

```python
conv_model.fit(
    X_train,
    y_train,
    batch_size=32,
    epochs=10,
    shuffle=True
)
```

```python
conv_model.evaluate(X_test, y_test,batch_size = 1000)
```

```python
CLASSES = np.array(['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck'])

preds = model.predict(X_test)
preds_single = CLASSES[np.argmax(preds,axis=-1)]
actual_single = CLASSES[np.argmax(y_test,axis=-1)]
```

```python
n_to_show = 10
indices = np.random.choice(range(0,len(X_test)), n_to_show)

fig = plt.figure(figsize=(15,3))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

for i,idx in enumerate(indices):
  img = X_test[idx]
  ax = fig.add_subplot(1, n_to_show, i+1)
  ax.axis('off')
  ax.text(0.5, -0.35, 'pred = ' + str(preds_single[idx]), fontsize=10, ha='center', transform=ax.transAxes)
  ax.text(0.5, -0.7, 'act = ' + str(actual_single[idx]), fontsize=10, ha='center', transform=ax.transAxes)
  ax.imshow(img)
```
